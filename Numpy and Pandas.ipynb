```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
peso = np.random.normal(70, 3, 1000)
altura = np.random.normal(1.68, .1, 1000)
```

```python
peso_np = np.array(peso)
altura_np = np.array(altura)
bmi_np = peso_np / (altura_np * altura_np)
print (peso_np, altura_np, bmi_np)
```

```python
peso_pd = pd.Series(peso)
altura_pd = pd.Series(altura)
```

```python
bmi = peso_pd / altura_pd*altura_pd
bmi
```
