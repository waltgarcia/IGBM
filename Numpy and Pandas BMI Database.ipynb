# Importamos librerias a utilizar

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
peso = np.random.normal(70, 3, 1000)
altura = np.random.normal(1.68, .1, 1000)
edad = np.random.random_integers(30, 60, 1000)
sex = np.random.randint(2, size=1000)
bmi = np.array(peso/(altura*altura))
```

# Verificamos tipo de variables 



```python
type(peso)
type(altura)
type(edad)
type(sex)
type(bmi)
```

```python
edad = np.random.random_integers(30, 60, 1000)
edad
```

```python
peso_np = np.array(peso)
altura_np = np.array(altura)
edad_np = np.array(edad)
bmi_np = peso_np / (altura_np * altura_np)
print (peso_np, altura_np, bmi_np)
```

```python
sex = np.random.randint(2, size=1000)
sex_pd = pd.Series(sex)
sex_pd
```

```python
peso_pd = pd.Series(peso)
altura_pd = pd.Series(altura)
edad_pd = pd.Series(edad)
```

# Armar DataFrame con Pandas

```python
together = pd.DataFrame({'Peso':peso_pd, 'Altura':altura_pd, 'IMC':peso_pd / (altura_pd * altura_pd), 'Edad': edad_pd, "Sex": sex})
together
```

```python
together.shape
```

```python
plt.scatter(peso_pd, altura_pd, facecolor= 'g')
plt.title('Relacion de IMC')
plt.xlabel('Peso')
plt.ylabel('Estatura')
plt.tight_layout()
plt.show()
```
